## Creating a little game

```python
def who_wins(a,b):
    print(poke["Name"][a] + " and " + poke["Name"][b] + " are fighting!!!")
    print(poke["Name"][a] + "'s power: " + str(poke["Total"][a]))
    print(poke["Name"][b] + "'s power: " + str(poke["Total"][b]))
    if poke["Total"][a] > poke["Total"][b]:        
        return print(poke["Name"][a] + " wins!")
    elif poke["Total"][a] < poke["Total"][b]:
        return print(poke["Name"][b] + " wins!")
    else:
        return print("It's a tie!")
```

```python
who_wins(384,502)
```

## Filtering data

```python
x = poke.loc[(poke['Type 1'] == 'Fire') & (poke["Attack"]>=100) & (poke["Legendary"] == False)]
x
```

```python
new_poke_df = poke.loc[(poke["Generation"] == 1) & (poke["Attack"]>=100)]
new_poke_df.reset_index(drop=True, inplace=True)
new_poke_df
```

```python
## ~ sign means "don't include"
poke.loc[~poke["Name"].str.contains("Mega")]
```

```python
## regular expressions
import re
poke.loc[poke["Name"].str.contains("^ch[a-z]", flags=re.I, regex=True)].reset_index()
```

## Conditional changes

```python
poke.loc[poke["Total"] > 600, ["Generation","Legendary"]] = ["Test 1", "Test 2"] 
poke
```

## Aggregate Statistics(groupby)

```python
df = pd.read_csv("modified_pokemon_data.csv")
```

```python
df.groupby(["Type 1"]).mean().sort_values("Attack", ascending=False)
```

```python
df["count"] = 0
df.groupby(["Type 1","Type 2"]).count()["count"]
```
